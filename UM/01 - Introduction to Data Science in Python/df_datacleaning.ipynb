```python
import pandas as pd

df = pd.read_csv('csv./presidents.csv')
df.head()
```

```python
df['First'] = df['President']
df['First'] = df['First'].replace('[ ].*', '', regex=True)
df.head()
```

```python
del(df['First'])

def splitname(row):
    row['First'] = row['President'].split(' ')[0]
    row['Last'] = row['President'].split(' ')[-1]
    return row

df = df.apply(splitname, axis='columns')
df.head()
```

```python
del(df['First'])
del(df['Last'])

pattern='(^[\w]*)(?:.* )([\w]*$)'
df['President'].str.extract(pattern).head()
```

```python
pattern='(?P<First>^[\w]*)(?:.* )(?P<Last>[\w]*$)'
names = df['President'].str.extract(pattern).head()
names
```

```python
df['First'] = names['First']
df['Last'] = names['Last']
df.head()
```

```python
df['Born'] = df['Born'].str.extract('([\w]{3} [\w]{1,2}, [\w]{4})')
df['Born'].head()
```

```python
df['Born'] = pd.to_datetime(df['Born'])
df['Born'].head()
```

```python
df['Died'] = df['Died'].str.extract('([\w]{3} [\w]{1,2}, [\w]{4})')
df['Died'].head()
```

```python
df['Died'] = pd.to_datetime(df['Died'])
df['Died'].head()
```

```python
df.head()
```

```python
df = df.rename(columns={'Age atstart of presidency': 'Age at start of presidency',
                        'Age atend of presidency': 'Age at end of presidency'})
df.head()
```

```python
df['Date start of presidency'] = df['Age at start of presidency'].str.extract('(?:[\w]{2}\syears, [\w]{1,3}\sdays)(.*)')
df['Date start of presidency'] = pd.to_datetime(df['Date start of presidency'])
```

```python
df['Date at end of presidency'] = df['Age at end of presidency'].str.extract('(?:[\w]{2}\syears, [\w]{1,3}\sdays)(?:\s?)([\w]{3} [\w]{1,2}, [\w]{4})')
df['Date at end of presidency'] = pd.to_datetime(df['Date at end of presidency'])
```

```python
df['Age at start of presidency'] = df['Age at start of presidency'].str.extract('([\w]{2}\syears, [\w]{1,3}\sdays)')
df['Age at start of presidency'].head()
```

```python
df.head()
```

```python
df['Age at end of presidency'] = df['Age at end of presidency'].str.extract('([\w]{2}\syears, [\w]{1,3}\sdays)')
df['Age at end of presidency'].head()
```
