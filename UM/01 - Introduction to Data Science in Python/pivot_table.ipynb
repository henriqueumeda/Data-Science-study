```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('csv/cwurData.csv')
df.head()
```

```python
def classify_university_world_rank(row):
    data = row['world_rank']
    if data <= 100:
        row['Rank_Level'] = 'First Tier Top University'
    elif data <= 200:
        row['Rank_Level'] = 'Second Tier Top University'
    elif data <= 300:
        row['Rank_Level'] = 'Third Tier Top University'
    else:
        row['Rank_Level'] = 'Other Top University'
    return row
```

```python
df = df.apply(classify_university_world_rank, axis='columns')
```

```python
# Outra forma de resolver essa questão
def create_category(ranking):
    if (ranking >= 1) & (ranking <= 100):
        return 'First Tier Top University'
    elif (ranking >= 101) & (ranking <= 200):
        return 'Second Tier Top University'
    elif (ranking >= 201) & (ranking <= 300):
        return 'Third Tier Top University'
    else:
        return 'Other Top University'

df['Rank_Level'] = df['world_rank'].apply(lambda x: create_category(x))
df.head()
```

```python
df.pivot_table(values='score', index='country', columns='Rank_Level', aggfunc=[np.mean]).head()
```
