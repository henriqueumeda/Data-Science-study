# Build a song recommender system

```
import turicreate
```

# Load some music data

```
song_data = turicreate.SFrame('/content/gdrive/My Drive/Turicreate/Week 5/song_data.sframe/')
```

# Explore our data

```
song_data
```

## Show the most popular songs in the dataset

```
song_data['song'].show()
```

# Count the number of unique users in the data

```
users = song_data['user_id'].unique()
```

```
len(users)
```

# Create a song recommender

```
train_data,test_data = song_data.random_split(.8,seed=0)
```

## Create a very simple popularity recommender

```
popularity_model = turicreate.popularity_recommender.create(train_data,
                                                           user_id = 'user_id',
                                                           item_id = 'song')
```

## Use the popularity model to make some predictions

```
popularity_model.recommend(users=[users[0]])
```

```
popularity_model.recommend(users=[users[1]])
```

# Build a recommender with personalization

```
personalized_model = turicreate.item_similarity_recommender.create(train_data,
                                                                  user_id = 'user_id',
                                                                  item_id = 'song')
```

## Apply personalized model to make song recommendations

```
personalized_model.recommend(users=[users[0]])
```

```
personalized_model.recommend(users=[users[1]])
```

# Apply model to find similar songs in the data set

```
personalized_model.get_similar_items(['With Or Without You - U2'])
```

```
personalized_model.get_similar_items(['Chan Chan (Live) - Buena Vista Social Club'])
```

# Compare the models quantitatively
We now formally compare the popularity and the personalized models using precision-recall curves. 

```
model_performance = turicreate.recommender.util.compare_models(test_data, [popularity_model, personalized_model], user_sample=.05)
```

The table shows that the personalized model provides much better performance.

# Recommending songs assignment

## 1. Counting unique users

```
kanye_west_users = song_data[song_data['artist'] == 'Kanye West']['user_id'].unique()
foo_fighters_users = song_data[song_data['artist'] == 'Foo Fighters']['user_id'].unique()
taylor_swift_users = song_data[song_data['artist'] == 'Taylor Swift']['user_id'].unique()
lady_gaga_users = song_data[song_data['artist'] == 'Lady GaGa']['user_id'].unique()
```

```
print('Number of unique users who listen to Kanye West: {}'.format(len(kanye_west_users)))
print('Number of unique users who listen to Foo Fighters: {}'.format(len(foo_fighters_users)))
print('Number of unique users who listen to Taylor Swift: {}'.format(len(taylor_swift_users)))
print('Number of unique users who listen to Lady GaGa: {}'.format(len(lady_gaga_users)))
```

## 2. Using groupby-aggregate to find the most popular and least popular artist

```
df_artist_listen_count = song_data.groupby(key_column_names='artist', operations={'total_count': turicreate.aggregate.SUM('listen_count')})
df_artist_listen_count.sort('total_count', ascending=False)
```

```
df_artist_listen_count.sort('total_count', ascending=True)
```

## 3. [OPTIONAL] Using groupby-aggregate to find the most recommended songs

```
subset_test_users = test_data['user_id'].unique()[0:10000]
```

```
user_recommend = personalized_model.recommend(subset_test_users,k=1)
```

```
user_recommend.groupby(key_column_names='song', operations={'count': turicreate.aggregate.COUNT()}).sort('count', ascending=False)
```
